```python

import openai
from openai import AzureOpenAI


deployment_name='gpt-4o' #This will correspond to the custom name you chose for your deployment when you deployed a model. 

client = AzureOpenAI(
    azure_endpoint = "ENDPOINT", 
    api_key="KEY",
    api_version="2025-01-01-preview")

# load instructions MD file
with open("quantum.md", "r") as file:
    instructions = file.read()

#parse file header to extract system instructions
if instructions.startswith("---"):
    end_header_index = instructions.find("---", 3)
    system_instructions = instructions[3:end_header_index].strip()
    instructions = instructions[end_header_index + 3:].strip()
else:
    system_instructions = ""
    instructions = ""

# extract system instructions individually
if system_instructions:
    print("System Instructions:")
    print(system_instructions)
    name = system_instructions.splitlines()[0].split(":")[1].strip()
    domain = system_instructions.splitlines()[1].split(":")[1].strip()
    tone = system_instructions.splitlines()[2].split(":")[1].strip()
    prompt = system_instructions.splitlines()[3].split(":")[1].strip()

# Load prompt template
with open("prompt_template.md", "r") as file:
    prompt_template = file.read()

#replace placeholders in the prompt template
prompt_instruction = prompt_template.replace("{{TONE_STYLE}}", tone).replace("{{DOMAIN_NAME}}", domain)

combinedContent = "{promptTemplate}\n\n# Knowledge Base\n{content}".format(promptTemplate=prompt_instruction, content=instructions)

messages=[]
messages.append({"role": "system", "content": combinedContent})
print(prompt)
print(f"example: What can you tell me about quantum computing?")

while True:
    user_input = input("User: ")
    if user_input.lower() in ['exit', 'quit']:
        break
    messages.append({"role": "user", "content": user_input})
    
    response = client.chat.completions.create(
        model=deployment_name,
        messages=messages
    )

    assistant_message = response.choices[0].message.content
    print(f"Assistant: {assistant_message}")
    messages.append({"role": "assistant", "content": assistant_message})
```
